### Проверка условий

Чтобы проанализировать кампании на выполнение условий (CPL менее 300, стоимость назначения менее 1100 и ДДР < 20%), сгруппируем данные в таблицах по id объявлений.   
Начнем с таблицы по кампаниям.

#### Группировка таблицы rk

В результате получим сгруппированную таблицу rk_group

```python
rk_group = rk.groupby(['campaign_id','ad_id']).agg({'impressions':'sum', 'clicks':'sum', 'spents':'sum'}).reset_index()
rk_group
```

Группировка прошла успешно, теперь по каждому объявлению можно посмотреть общее количество показов, кликов и потраченную сумму.

#### Группировка таблицы ld

##### Создадим новый столбец в таблице 'ld'

Столбец 'purpose' будет отображать, является ли действие - назначением. Он нужен, чтобы посчитать количество целевых действий.

```python
def experience(row):
    status = row['status']
    if status == 'Оплата' or status == 'Д назначена' or status == 'Д состоялась':
        return True
    return False
```

```python
ld['purpose'] = ld.apply(experience, axis = 1)
ld.head()
```

Получилось. Теперь сгруппируем таблицу по 'utm_content'.

##### Группировка таблицы

Сгруппированную таблицу сохраним в переменную 'ld_group'

```python
ld_group = ld.groupby('utm_content').agg({'status':'count','purpose':'sum', 'price':'sum'}).reset_index()
ld_group.sort_values('status', ascending=False)
```

Отлично, теперь можно соединить таблицы.

#### Объединение таблиц

Для удобства переименуем столбцы в таблице ld_group

```python
ld_group.columns = ['ad_id', 'leads', 'purpose_cnt', 'revenue']
ld_group.head()
```

Теперь объединим таблицы по общему полю - id объявления.

```python
df = rk_group.merge(ld_group, on='ad_id', how='left')
df
```

Объединение прошло успешно, в новой таблице 1866 строк.

#### Группировка итоговой таблицы по кампаниям

```python
df_group = df.groupby('campaign_id').agg({'impressions':'sum',
                                          'clicks':'sum',
                                          'spents':'sum',
                                          'leads':'sum',
                                          'purpose_cnt':'sum',
                                          'revenue':'sum'})
df_group
```

Таблица сгруппирована, добавим столбцы с подсчетом нужных нам метрик.

```python
df_group['CPL'] = round(df_group['spents'] / df_group['leads'], 2)
df_group['cost_purpose'] = round(df_group['spents'] / df_group['purpose_cnt'], 2)
df_group['DPP'] = round(df_group['spents'] / df_group['revenue'], 4)

df_group.sort_values('DPP')
```

Столбцы успешно добавлены.


#### Проверка условий

Теперь проверим кампании на условия: 
- CPL менее 300, 
- стоимость назначения менее 1100 
- и ДДР < 20%

Выведем только те кампании, которые соответствуют данным условиям.

```python
df_campaign = df_group.query('CPL <= 300 and cost_purpose <= 1100 and DPP <= 0.2')
df_campaign
```

```python
print('Количество кампаний, который соответствуют условиям -', len(df_campaign))
df_campaign.index
```

15 кампаний, который соответствуют условиям - 85469607, 85578565, 85607658, 85699073, 85699074, 85852144, 85858138, 85912106, 86111338, 86375282, 86375284, 86540275, 87554731, 87594336, 87929708 .  
Из них минимальный CPL и минимальная цена назначения у кампании - 87929708 (CPL=150,05р, цена назначения=262.59), а самая минимальная доля расходов на рекламу у кампаний 87554731,	86375284 и 87929708 (менее 5%).
